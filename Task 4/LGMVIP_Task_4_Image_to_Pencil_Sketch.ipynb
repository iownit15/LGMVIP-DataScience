# Importing the Libraries

```python
import cv2
from PIL import Image
from IPython.display import display
from google.colab.patches import cv2_imshow
```

# Importing and the Image and resizing

```python
img = cv2.imread('scene.jpg', 1)
img = cv2.resize(img, (720, 480))
```

# Converting color order from BRG to RGB since openCV uses BRG

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

# Converting the RGB image to grayscale

```python
gscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gscale))
```

# Inverting the image colors

```python
invImg = cv2.bitwise_not(gscale)
display(Image.fromarray(invImg))
```

# Softening the image

```python
softImg = cv2.GaussianBlur(invImg, (5, 5), 0, 0)
display(Image.fromarray(softImg))
```

# Inverting the softened image

```python
softInv = 255 - softImg
finalSketch = cv2.divide(gscale, softInv, scale=256.0)
display(Image.fromarray(finalSketch))
```

# Comparison of Original Image and Pencil Sketch

```python
display(Image.fromarray(img))
display(Image.fromarray(finalSketch))
```
